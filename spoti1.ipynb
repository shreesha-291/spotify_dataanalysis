```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_tracks=pd.read_csv("/content/tracks.csv")
df_tracks.head()
```

```python
pd.isnull(df_tracks).sum()
```

```python
df_tracks.info()
```

```python
sorted_df=df_tracks.sort_values("popularity",ascending=False)
sorted_df.head(10)
```

```python
df_tracks.describe().transpose()
```

```python
most_popular=df_tracks.query('popularity>50',inplace=False).sort_values('popularity',ascending=False)
most_popular[:10]
```

```python
#changing index to release date
```

```python
df_tracks.set_index("release_date",inplace=True)
df_tracks.index=pd.to_datetime(df_tracks.index)
df_tracks.head()
```

```python
df_tracks[["artists"]].iloc[18]
```

```python
# convert duration in millisecond to just second
```

```python
df_tracks["duration"]=df_tracks["duration_ms"].apply(lambda x: round(x/1000))
df_tracks.drop("duration_ms",axis=1,inplace=True)
```

```python
df_tracks.duration.head()
```

```python
# visualisation starts from here
```

```python
corr_df=df_tracks.drop(["key","mode","explicit"],axis=1).corr(method="pearson")
plt.figure(figsize=(14,6))
heatmap=sns.heatmap(corr_df,annot=True,fmt=".1g",vmin=-1,vmax=1,center=0,cmap="rocket",linewidths=1,linecolor="black")
heatmap.set_title("correlation heatmap between variable")
heatmap.set_xticklabels(heatmap.get_xticklabels(),rotation=90)
```

```python
#making sample dataset
```

```python
sample_df=df_tracks.sample(int(0.004*len(df_tracks)))
print(len(sample_df))
print(len(df_tracks))
```

```python
#regression plot loudness and energy
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data=sample_df,y="loudness",x="energy",color="c").set(title="LOUDNESS VS ENERGY CORRELATION")
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data=sample_df,y="popularity",x="acousticness",color="b").set(title="POPULARITY VS ACOUSTICNESS CORRELATION")
```

```python
#distribution table for knowing no of songs done since 1922
```

```python
df_tracks['dates']=df_tracks.index.get_level_values('release_date')
df_tracks.dates=pd.to_datetime(df_tracks.dates)
years=df_tracks.dates.dt.year
```

```python
sns.displot(years,discrete=True,aspect=2,height=5,kind="hist").set(title="NUMBER OF SONGS PER YEAR")
```

```python
total_dr=df_tracks.duration
fig,ax=plt.subplots(figsize=(18,20))
fig=sns.barplot(y=total_dr,x=years,ax=ax,errwidth=False).set(title="YEAR VS DURATION")
plt.xticks(rotation=90)
```

```python
#same now taken with  line plot
```

```python
total_dr=df_tracks.duration
sns.set_style(style="whitegrid")
fig,ax=plt.subplots(figsize=(18,20))
fig=sns.lineplot(y=total_dr,x=years,ax=ax).set(title="YEAR VS DURATION")
plt.xticks(rotation=60)
```
